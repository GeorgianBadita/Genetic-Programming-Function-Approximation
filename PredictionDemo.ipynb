```python
"""
IMPORTS
"""
from genetic_algorithm.population import *
from genetic_algorithm.ga_operations import *
from genetic_algorithm.algorithm import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

<b>Algorithm parameters</b>

```python
"""
ALGORITHM PARAMETERS
"""

SIZE = 1 # number of variables
MAX_DEPTH = 20 #maximum depth of a tree 
FUNCTIONS = {1: ['sin','cos','e','ln','tg','tanh','abs'], 2:['+', '-', '*', '/']} #unary and binary functions
TERMINAL_SET = ['x'+str(i) for i in range(SIZE)] #create parameters name
DEPTH = 4 #initial depth of trees
POP_SIZE = 1000 #number of chromosomes in population
NUM_FOR_SELECTION = 1000//3 #number of chromosomes to be chosen for selection
NUMBER_ITERS = 1000 #number of iterations
```

<b>Our first predicted function</b>

```python

"""
DEFINE A FUNCTION TO PREDICT
"""
def f(x):
    return x**2 + 2*x

X = [[x] for x in np.arange(0, 10, 0.01)] #function inputs
y = [[f(x[0])] for x in X] #function outputs

pop = Population(POP_SIZE, NUM_FOR_SELECTION, FUNCTIONS, TERMINAL_SET, 6, MAX_DEPTH) #create the population
alg = Algorithm(pop, NUMBER_ITERS, X, y, epoch_feedback=500)  #create the algorithm
best = alg.train() #train the alrogithm
```

<b>Graphs are overlaping since we exactly approximated the function</b>

```python
print(best.gen) #print best function
print(best.fitness)

y_pred = [[best.evaluate_arg(x)] for x in X] #predictions of our best calculated function
 
plt.plot(X, y, color='r', dashes=[6, 2], label='Expected')  #plot original function 
plt.plot(X, y_pred, color='g', dashes=[6, 3], label='Predicted') #plot calculated function
plt.legend()
plt.show() #show the plot
```
